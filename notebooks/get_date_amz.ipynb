```python
import os
from simplecube import cube_query, simple_cube, cube_get_data
```

```python
path_dir = os.path.abspath("")
```

```python
datacube=cube_query(
    collection="AMZ1-WFI-L4-SR-1",
    start_date="2024-06-01",
    end_date="2025-05-31",    
    bbox="-52.4048,-24.6221,-51.8005,-24.2169",
    bands=["BAND3","BAND4"]
)
```

```python
#cube_get_data(datacube)
```

```python
cube = simple_cube(
    data_dir=os.path.join(path_dir),
    datacube=datacube,
    source='bdc-amz',
    band='BAND3'
)

cube
```

```python
cube_2d = cube['band_data'].sel(time='2024-06-11')

cube_2d.plot()
```
