```python
from simplecube import simple_cube
```

```python
stac_url = "https://data.inpe.br/bdc/stac/v1"
```

```python
S2_cube=simple_cube(
    stac_url=stac_url,
    collection="S2-16D-2",
    start_date="2024-01-01",
    end_date="2024-12-31",
    bbox="-45.5006,-13.0066,-45.0172,-12.6062",
    bands=["B03","B04","B08"]
)

S2_cube
```

```python
nir_band = S2_cube.nir
red_band = S2_cube.red
green_band = S2_cube.green

ndvi = (nir_band - red_band)/(nir_band + red_band)
evi2 = 2.4 * (nir_band - red_band) / (nir_band + red_band + 1.0)
ndwi = (green_band - nir_band)/(green_band + nir_band)
savi = ((nir_band - red_band) / (nir_band + red_band + 0.5)) * (1 + 0.5)

S2_cube['ndvi'] = ndvi
S2_cube['evi2'] = evi2
S2_cube['ndwi'] = ndwi
S2_cube['savi'] = savi

S2_cube
```
