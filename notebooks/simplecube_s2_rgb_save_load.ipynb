```python
from simplecube import simple_cube, save_xarray
```

```python
stac_url = "https://data.inpe.br/bdc/stac/v1"
```

```python
S2_RGB_cube=simple_cube(
    stac_url=stac_url,
    collection="S2-16D-2",
    start_date="2024-01-01",
    end_date="2024-12-31",
    bbox="-46.7743,-15.9686,-46.5824,-15.7992",
    bands=["B02","B03","B04"]
)

S2_RGB_cube
```

```python
save_xarray(S2_RGB_cube, 'my_data.nc')
```

```python
from simplecube import load_xarray

S2_RGB_cube = load_xarray('my_data.nc')

S2_RGB_cube
```
