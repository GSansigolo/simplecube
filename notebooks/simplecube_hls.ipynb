```python
import os
from simplecube import local_simple_cube
```

```python
path_dir = os.path.abspath("")
```

```python
NDVI_cube = local_simple_cube(
    data_dir=os.path.join(path_dir),
    collection="HLS.v2.0",
    bands=["NDVI"],
    tile="T23LCC",
    bbox="-46.8292,-16.0656,-46.3692,-15.7169",
    source='nasa'
)

NDVI_cube
```
