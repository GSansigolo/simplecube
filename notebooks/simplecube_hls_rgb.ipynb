```python
import os
from simplecube import local_simple_cube
```

```python
path_dir = os.path.abspath("")
```

```python
HLS_cube = local_simple_cube(
    collection="HLS.v2.0",
    data_dir=os.path.join(path_dir),
    source='nasa',
    bands=["B02","B03","B04"],
    tile="T23LMG",
    bbox="-45.5,-12.8,-44.8,-12.1"
)

HLS_cube
```
